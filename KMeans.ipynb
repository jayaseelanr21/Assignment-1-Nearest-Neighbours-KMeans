## Cluster similar Houses


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("house_rental_data.csv.txt",sep=",")
data.head()
```

```python
dic = {'Unnamed: 0':'Id'}
data = data.rename(dic,axis = 1)
data.head()
```

```python
data = data.set_index('Id')
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
x = data.drop('Price',axis = 1)
y = data['Price']
```

```python
x
```

## KMeans Clustering

```python
from sklearn.cluster import KMeans
kmeans = KMeans()
kmeans
```

## Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
mms
```

```python
x = mms.fit_transform(x)
x
```

```python
kmeans.fit(x)
```

```python
kmeans.cluster_centers_
```

```python
cluster_label = kmeans.fit_predict(x)
cluster_label
```

```python
df = pd.DataFrame(cluster_label, columns= ['Cluster'])
```

```python
df = mms.fit_transform(df)
df
```
